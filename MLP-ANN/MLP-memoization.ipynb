```python
#  fib

def fib(n):
    if n ==1 or n ==0:
        return 1
    else:
        return fib(n-1) + fib(n-2)
    
    #  This is inefficent code
```

```python
fib(6)
```

```python
fib(36)
```

```python
# As this code O(n) = 2^n to solve this we use memoization
```

```python
#  using memoization
def fib(n,d):
    if n in d:
        return d[n]
    else:
        d[n] = fib(n-1,d) + fib(n-2,d)
        return d[n]
    
```

```python
d = {0:1,1:1}
fib(2978,d)
```
