```python
import tensorflow 
from tensorflow import keras 
from keras import Sequential 
from keras.layers import Dense
```

```python
model =Sequential()

model.add(Dense(10,activation='relu', input_dim=2, kernel_initializer='he_normal')) # in every layer we need to set the kernel_initializer in default it is always glorot_uniform, u can change it into glorot_normal and he_uniform too
```
