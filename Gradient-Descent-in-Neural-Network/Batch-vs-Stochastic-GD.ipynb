```python
import numpy as np 
import pandas as pd 
import time
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df = df[['Age','EstimatedSalary','Purchased']]
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split as tts 
X_train,X_test,y_train,y_test = tts(X_scaled,y,test_size=0.2,random_state=2)
```

```python
import tensorflow as tf 
from tensorflow import keras
from keras import Sequential 
from keras.layers import Dense 
```

```python
model = Sequential()

model.add(Dense(10,activation='relu',input_dim=2))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', metrics=['accuracy'])
# batch_size=1 --> Stochastic, batch_size=X.shape[0] --> Batch GD
hist = model.fit(X_train,y_train,epochs=10,batch_size=1)

# 
```

```python
model1 = Sequential()

model1.add(Dense(10,activation='relu',input_dim=2))
model1.add(Dense(10,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.summary()
```

```python
model1.compile(loss='binary_crossentropy', metrics=['accuracy'])

history = model1.fit(X_scaled,y,epochs=400, batch_size=400, validation_split=0.2)
```

```python
X_scaled.shape
```

```python
# Stochastic GD gives higher accuracy but takes more time to train the model 
#   And also convergence will give effecient result with least epochs
```
