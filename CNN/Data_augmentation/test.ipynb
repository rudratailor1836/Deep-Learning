```python
from keras.preprocessing import image 
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('D:/Deep-Learning/CNN/Data_augmentation/28.jpg')
```

```python
import matplotlib.pyplot as plt 

plt.imshow(img)
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    width_shift_range=0.2,
    height_shift_range=0.2,
    fill_mode='constant' # reflect,nearest,constant
)
```

```python
img = image.img_to_array(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,500,499,3)
```

```python
i=0
```

```python
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='D:\Deep-Learning\CNN\Data_augmentation'):
    i = i+1

    if i == 10:
        break


# flow and flow_from_dir both are for 1 single img and more than 1 images in a directory respectively
```
